## 1. `pypdf` (PDF 파일용)

```python
%pip install -q pypdf
```

```python
import pypdf
import os

def load_pdf_with_pypdf(file_path: str) -> str:
    """pypdf를 사용하여 PDF 파일에서 텍스트를 추출한다."""
    text = ""
    with open(file_path, "rb") as f:
        reader = pypdf.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text
```

```python
pdf_path = "../dataset/AI_Understanding.pdf"
document_text = load_pdf_with_pypdf(pdf_path)
print(f"--- pypdf 추출 결과 ---\n{document_text}")
```

## 2. fitz (PDF 파일용)

```python
%pip install -q fitz
```

```python
import fitz  # PyMuPDF
import os

def load_pdf_with_fitz(file_path: str) -> str:
    """PyMuPDF(fitz)를 사용하여 PDF 파일에서 텍스트를 추출한다."""
    text = ""
    doc = fitz.open(file_path)
    for page in doc:
        text += page.get_text() + "\n"
    doc.close()
    return text
```

```python
pdf_path = "../dataset/AI_Understanding.pdf"
document_text = load_pdf_with_fitz(pdf_path)
print(f"--- fitz 추출 결과 ---\n{document_text}")
```

## 3. `python-docx` (Word 파일용)

```python
%pip install -q python-docx
```

```python
from docx import Document
import os

def load_docx(file_path: str) -> str:
    """python-docx를 사용하여 .docx 파일에서 텍스트를 추출한다."""
    doc = Document(file_path)
    full_text = [para.text for para in doc.paragraphs]
    return "\n".join(full_text)
```

```python
docx_path = "../dataset/AI_Understanding.docx"
document_text = load_docx(docx_path)
print(f"--- python-docx 추출 결과 ---\n{document_text}")
```

## 4. `pandas` (Excel, CSV 파일용)

```python
%pip install -q openpyxl
```

```python
import pandas as pd
import os

def load_excel(file_path: str) -> str:
    """pandas를 사용하여 Excel 파일의 모든 시트 내용을 텍스트로 변환한다."""
    xls = pd.ExcelFile(file_path)
    full_text = ""
    for sheet_name in xls.sheet_names:
        df = pd.read_excel(xls, sheet_name=sheet_name)
        if not df.empty:
            full_text += f"--- Sheet: {sheet_name} ---\n"
            full_text += df.to_string(index=False) + "\n\n"
    return full_text

def load_csv(file_path: str) -> str:
    """pandas를 사용하여 CSV 파일 내용을 텍스트로 변환한다."""
    df = pd.read_csv(file_path)
    return df.to_string(index=False)
```

```python
excel_path = "../dataset/advertising.xlsx"
csv_path = "../dataset/advertising.csv"

excel_text = load_excel(excel_path)
print(f"--- pandas Excel 추출 결과 ---\n{excel_text}")

csv_text = load_csv(csv_path)
print(f"--- pandas CSV 추출 결과 ---\n{csv_text}")
```
